```python
import pandas as pd
import numpy as np
import tensorflow as tf
import PIL 
import os
import matplotlib.pyplot as plt
```

```python
final_model = tf.keras.models.load_model("transfer_learning_model.h5")
```

```python
def predict(image_path, model):
    img = PIL.Image.open(image_path) # image extension *.png,*.jpg
    new_width, new_height  = 256, 256
    img = img.resize((new_width, new_height), PIL.Image.ANTIALIAS)
    photo_check =tf.expand_dims(np.array(img), axis=0)
    return model.predict(photo_check).round()[0]
```

```python
examples_dir_path = "data/application_photo_examples/"
photo_paths = os.listdir(examples_dir_path)
print(photo_paths)
```

```python
for photo in photo_paths:
    image_path = os.path.join(examples_dir_path, photo)
    prediction = predict(image_path, final_model)
    if prediction == 0:
        print("clean")
    else:
        print("dirty")
        
    plt.imshow(PIL.Image.open(image_path))
    plt.show()
```
